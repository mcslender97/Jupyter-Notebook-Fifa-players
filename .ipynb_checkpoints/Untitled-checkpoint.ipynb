```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('players_20.csv')
df
```

To show all column names, list(df) can be used to list column info

```python
list(df)
```

```python
# Show number of row and columns. df.info shows information of the data frame
df.info()
```

```python
# show number of players and their countries
df['sofifa_id'].count()
```

```python
#show countries
df['nationality'].unique()
```

```python
#show nationalities and num of players
df.groupby(['nationality'])[['sofifa_id']].count()
```

```python
df.groupby(['nationality'])['sofifa_id'].count().sort_values(ascending=False).head(10)
```

```python
# create bar plot of top 5 countries
plot_df = df.groupby(['nationality'])[['sofifa_id']].count().sort_values(by=['sofifa_id'],ascending=False).head(5)
```

```python
no_of_players = plot_df['sofifa_id']
nationality_col = plot_df.index
plt.bar(nationality_col,no_of_players, color = 'green')
plt.xticks(rotation = 90)
plt.title("Countries with most players")
plt.xlabel("Countries")
plt.ylabel("No of Players")
plt.show()
```

```python
# top 5 players short name and wages
df[['short_name','wage_eur']].head(5)
```

```python
# top 5 players short name and wages w highest salaries
player_wage_df=df[['short_name','wage_eur']].sort_values(by=['wage_eur'],ascending=False).head(5)
player_wage_df
```

```python
# Bar plot of 8)
salary = player_wage_df['wage_eur']
players_col = player_wage_df['short_name']
plt.bar(players_col,salary)
plt.xticks(rotation = 90)
plt.title("Top 5 players based on Salary")
plt.xlabel("Players name")
plt.ylabel("Wage in EUR")
plt.show()
```

```python
# top 10 records of Germany
df[df["nationality"]=="Germany"].head(10)
```

```python
# top 5 records of Germany who have max height, weight and wage
df[df["nationality"]=="Germany"].sort_values(by=['height_cm','weight_kg','wage_eur'],ascending=False).head(5)
```

```python
#shortname and wage of top 5 Germnay players
df.loc[df["nationality"]=="Germany",["short_name","wage_eur"]].head(5)
```

```python
#top 5 players by shooting skills
df.sort_values(by=['shooting'],ascending=False).head(5)[["short_name"]]
```
